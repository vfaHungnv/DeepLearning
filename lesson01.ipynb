Đề bài:
Cho phuong trình y=ax+b
- Cho nghiệm x=[x1,x2,....,xn], y=[y1,y2,....,yn]
- Tìm w (weight)
- (b: bias)

```python
import os
os.environ["CUDA_VISIBLE_DEVICES"]="-1" #Ep training tren CPU
from keras.layers import *
from keras.models import *
from keras import optimizers
```

```python
x=[1, 2, 3, 4]
y=[2, 4, 6, 8]
x, y
```

```python
model=Sequential([
    Dense(1, input_shape=(1,))
])
model.summary()
```

```python
sgd=optimizers.SGD(lr=0.01) #Toc do traning
model.compile(loss='mse',optimizer=sgd)
```

```python
model.fit(x,y,epochs=700) #Training 700 lan
```

```python
y_pred=model.predict([11.])
y_pred
```
